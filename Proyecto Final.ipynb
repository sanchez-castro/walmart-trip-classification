### Feature Engineering

```python
#Creando dummies de los dias
def feature_engineering(data):
    "Funcion para transformas algunas de las variables en los datasets"
    #Creamos dummies de los departamentos
    data = pd.get_dummies(data, columns = ["DepartmentDescription"])
    #Pasamos los dias a factores ordenados
    dias = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    data.Weekday = data.Weekday.astype("category", ordered = True, 
                                       categories = dias).cat.codes    
    #Creando dummie especial si hubo algun retorno de producto
    data.loc[data.ScanCount < 0, 'Return'] = 1
    data.loc[data.Return != 1, 'Return'] = 0
    
    return data
```

```python
feature_engineering(train)
```
